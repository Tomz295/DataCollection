```python
def dataGen():
    while True:
        with open("Output.txt", "r") as file:
            while True:
                text = file.readline()
                if not text:
                    break
                yield text
```

```python
import time
a=0
ite = dataGen()
for i in range(450):
    item = next(ite)
    time.sleep(0.1)
    a += 1
    print(item)
    print(a)
```
